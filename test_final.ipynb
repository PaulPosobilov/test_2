```python
#Импортируем библиотку для работы
import pandas as pd
```

```python
#Читаем первый лист
df=pd.read_excel(r'C:\Users\Paul\Documents\ГКУ МЦРСТ\Python.xlsx',sheet_name='Прикрепленное население')
```

```python
#Смотрим информацию о листе
print(df.info())
```

Сформируем первые две требуемые таблицы

```python
#Формируем сводную таблицу в разбивке по округам и считаем количество прикрепленного населения по месяцам
table_1_1=df.groupby(['Округ']).agg({'Прикрепленное население на 01.07':sum,
                           'Прикрепленное население на 01.08':sum,
                           'Прикрепленное население на 01.09':sum}).copy()
table_1_1
```

```python
#Формируем две колонки с приростом приклепения населения в процентах с округлением до двух знаков после запятой
col_1= list(
    round(
        ((table_1_1['Прикрепленное население на 01.08']-table_1_1['Прикрепленное население на 01.07'])/
         table_1_1['Прикрепленное население на 01.07']).mul(100),2)
)
col_2= list(
    round(
        ((table_1_1['Прикрепленное население на 01.09']-table_1_1['Прикрепленное население на 01.08'])/
         table_1_1['Прикрепленное население на 01.08']).mul(100),2)
)
```

```python
#Добавляем новые столбцы
table_1_1.insert(2, "Прирост в % июль-август", col_1)
table_1_1.insert(4, "Прирост в % август-сентябрь", col_2)
table_1_1
```

```python
#Читаем второй лист, чтобы составить вторую таблицу с рейтингом специальностей по количеству приемов в сентябре
df2=pd.read_excel(r'C:\Users\Paul\Documents\ГКУ МЦРСТ\Python.xlsx',sheet_name='Структура записей и приемов')
```

```python
#Смотрим информацию о листе
df2.info()
```

Сформируем вторую таблицу

```python
#Делаем сводную таблицу специальностей врачей по количеству приемов в сентябре. Сортировку задаём от большего к меньшему.
table_1_2=df2.groupby('Специальность врача').agg({'Количество состоявшихся приемов, в т. ч. (Сентябрь)':sum}).astype('Int64')
table_1_2=table_1_2.sort_values(by='Количество состоявшихся приемов, в т. ч. (Сентябрь)',ascending=False)
table_1_2
```

Формируем третью таблицу

```python
#Читаем третий лист
df3=pd.read_excel(r'C:\Users\Paul\Documents\ГКУ МЦРСТ\Python.xlsx',sheet_name='Количество ставок')
```

```python
#Смотрим информацию о листе
df3.info()
```

```python
#Сделаем первые три колонки
table_3=df2.groupby(['Округ','Наименование поликлиники'])['Количество состоявшихся приемов, в т. ч. (Сентябрь)'].sum().reset_index()
table_3['Количество состоявшихся приемов, в т. ч. (Сентябрь)']=table_3['Количество состоявшихся приемов, в т. ч. (Сентябрь)'].astype('int64')
table_3.head(5)
```

```python
#Добавим колонку с долей внеплановых посещений
col_vneplan=list(
    round((df2.groupby(['Округ','Наименование поликлиники'])['количество внеплановых приемов (Сентябрь)'].sum()/
    df2.groupby(['Округ','Наименование поликлиники'])['Количество состоявшихся приемов, в т. ч. (Сентябрь)'].sum()).
                mul(100),2)
)
table_3['Доля внеплановых в %']=col_vneplan
table_3.head(5)
```

```python
#Добавим колонку с долей несостоявшихся плановых посещений
col_nesost=list(round(
    ((df2.groupby(['Округ','Наименование поликлиники'])['Количество записей (Сентябрь) '].sum()-
     df2.groupby(['Округ','Наименование поликлиники'])['количество плановых приемов (Сентябрь)'].sum())/
    df2.groupby(['Округ','Наименование поликлиники'])['Количество записей (Сентябрь) '].sum()).
                mul(100),2
))
table_3['Доля несост. план. %']=col_nesost
table_3.head(5)
```

Была обнаружена проблема в наименовании поликлини Городская поликлиника № 37 (лишний пробел в конце строки)

```python
#Удаляем на первом листе лишние пробелы возле строки Городская поликлиника № 37
df['Наименование поликлиники'] = df['Наименование поликлиники'].map(str.strip)
```

```python
#Проверка
df[df['Наименование поликлиники']=='Городская поликлиника № 37']
```

```python
#Создадим таблицу с первого листа, убрав детские поликлиники
discard = ["Детская"]
df_vzros= df[~df['Наименование поликлиники'].str.contains('|'.join(discard ))]
df_vzros
```

```python
#Добавим колонку с количеством прикрепленных на 01.09
col_prikrep=list(
    df_vzros.groupby(['Округ','Наименование поликлиники'])['Прикрепленное население на 01.09'].sum()   
)
table_3['Прикреп. 01.09']=col_prikrep
table_3.head(5)
```

```python
#Добавим колонку с количеством ставок на 01.09
df3.reset_index(drop=True)
col_stavok=list(
    df3.groupby(['Округ','Наименование поликлиники'])['Количество ставок по состоянию на 01.09.2021'].sum()   
)
col_stavok.insert(29, 0)
table_3['Ставок 01.09']=col_stavok
table_3.head(5)
```

Сформируем четвёртую таблицу

```python
# Сформируем первые три столбца
table_4=df_vzros.groupby(['Округ','Наименование поликлиники'])['Прикрепленное население на 01.08'].sum().reset_index()
table_4['Прикрепленное население на 01.08']=table_4['Прикрепленное население на 01.08'].astype('int64')
table_4.head(5)
```

```python
#Добавим колонку состоявишхся приёмов врача-невролога на 01.08
col_nev=list(df2[df2['Специальность врача']=='Врач-невролог'].groupby(
 ['Округ','Наименование поликлиники']   
)['Количество состоявшихся приемов, в т. ч. (Август) '].sum())
table_4['col_priemov_nevrolog']=col_nev
table_4['col_priemov_nevrolog']=table_4['col_priemov_nevrolog'].astype('int64')
table_4.head(5)
```

```python
#Добавим колонку со ставкой врача-невролога на 01.08
col_nev_stav=list(round(df3[df3['Специальность врача']=='Врач-невролог'].groupby(
 ['Округ','Наименование поликлиники']   
)['Количество ставок по состоянию на 01.08.2021'].sum(),2))
col_nev_stav.insert(29, 0)
table_4['col_stavok_nevrolog']=col_nev_stav

table_4.head(5)
```

```python
#Добавим колонку приёмов на ставку на 01.08
table_4['приёмов на ставку'] = round(table_4['col_priemov_nevrolog']/table_4['col_stavok_nevrolog'],0)
#Добавим колонку приёмов на 10 тыс. прикреп на 01.08
table_4['приёмов на 10 тыс. прикреп'] = round((table_4['col_priemov_nevrolog']/(table_4['Прикрепленное население на 01.08']/10000))).astype('int64')
#Добавим колонку ставок на 10 тыс. прикреп на 01.08
table_4['ставок на 10 тыс. прикреп'] = round(table_4['col_stavok_nevrolog']/(table_4['Прикрепленное население на 01.08']/10000),2)
```
